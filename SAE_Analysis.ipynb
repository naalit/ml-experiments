```python
import torch
from transformers import GPT2Tokenizer
```

```python
data = torch.load('feature_map_140000')
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
```

```python
yellow = '\033[93m'
green = '\033[92m'
red = '\033[91m'
blue = '\033[94m'
pink = '\033[95m'

def color(col):
    # pink -> red -> yellow -> green -> blue -> black/white (default)
    if col > 0.9:
        return pink
    if col > 0.8:
        return red
    if col > 0.6:
        return yellow
    if col > 0.4:
        return green
    if col > 0.2:
        return blue
    return "\033[0m"

def process(data):
    ret = []
    for (act, text, vect) in data:
        if act == 0.0:
            continue
        tokens = tokenizer.encode(text)
        buf = f"\033[0m{act}:\n"
        assert(len(tokens) == vect.shape[0])
        for i in range(len(tokens)):
            buf += color(vect[i].item() / act)
            buf += tokenizer.decode(tokens[i])
        ret.append(buf)
    return ret
```

```python
for j in range(20):
    print(f'---{j}---')
    for i in process(data[j]):
        print(i, '\n')
```
